```python
#function to explore neighbourhoods in downtown toronto
def getNearbyVenues(names, latitudes, longitudes, radius=500,categoryIds=''):
    
    venues_list=[]
    for name, lat, lng in zip(names, latitudes, longitudes):
        print(name)
            
        # create the API request URL
        url = 'https://api.foursquare.com/v2/venues/explore?&client_id={}&client_secret={}&v={}&ll={},{}&radius={}&limit={}'.format(
            CLIENT_ID, 
            CLIENT_SECRET, 
            VERSION, 
            lat, 
            lng, 
            radius, 
            LIMIT)
        
        if (categoryIds != ''):
               url = url + '&categoryId={}'
               url = url.format(categoryIds)        
            
        # make the GET request
        results = requests.get(url).json()["response"]['groups'][0]['items']
        
        # return only relevant information for each nearby venue
        venues_list.append([(
            name, 
            lat, 
            lng, 
            v['venue']['name'], 
            v['venue']['location']['lat'], 
            v['venue']['location']['lng'],  
            v['venue']['categories'][0]['name']) for v in results])

    nearby_venues = pd.DataFrame([item for venue_list in venues_list for item in venue_list])
    nearby_venues.columns = ['Neighbourhood', 
                  'Neighbourhood Latitude', 
                  'Neighbourhood Longitude', 
                  'Venue', 
                  'Venue Latitude', 
                  'Venue Longitude', 
                  'Venue Category']
    
    return(nearby_venues)
```

```python
#function to add new columns to the existing dataframe
def addcol(mod_df, newcoltitle, in_df):
    grouped_df = in_df.groupby('Neighbourhood').count()
    
    for i in mod_df['Neighbourhood']:
        try:
            mod_df.loc[mod_df['Neighbourhood'] == i,newcoltitle] = grouped_df.loc[i, 'Venue']
        except:
            mod_df.loc[mod_df['Neighbourhood'] == i,newcoltitle] = 0
```

```python
addcol(newdata, 'Pizza Place', eastyork_venues_pizza)
addcol(newdata, 'Schools', eastyork_venues_schools)
addcol(newdata, 'Offices', eastyork_venues_off)
addcol(newdata, 'Parks', eastyork_venues_park)
newdata
```

```python
#assigning weights to each of the places considered; pizza place, office, school and park
# If the venue is pizza place, we assign a negative weight as we want to avoid a place which already has too many pizza places
pizza_wt = -1

# if there are schools nearby, school students can be considered as good customers; so assign a positive weight
schools_wt = 1

# office employees can spend more than school kids; so little more wightage
off_wt = 1.5

# kids who come to park can spend much more as they often come with parents and it will be usually an outing
parks_wt = 2
```

```python
weighted_newdata = newdata[['Neighbourhood']].copy()
weighted_newdata
```

```python
weighted_newdata['Popularity Score'] = newdata['Pizza Place'] * pizza_wt + newdata['Schools'] * schools_wt + newdata['Offices'] * off_wt + newdata['Parks'] * parks_wt
weighted_newdata = weighted_newdata.sort_values(by=['Popularity Score'], ascending=False)
weighted_newdata
```
